## Dataset Preview Analysis and Data Cleaning

```python
# Converting Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df.sort_values('Date', inplace=True)
```

```python
df.head()
```

```python
# Checking for na values
missing = df.isna().sum()
missing_percent = (missing / len(df) * 100).round(2)
pd.DataFrame({'Missing Values': missing, 'Percent Missing': missing_percent})
```

```python
# Summary of Column and Dtypes
df.info()
```

```python
# Column statistics (Transposing for easier reading)
df.describe().T
```

```python
# Summary Statistics
df.select_dtypes(include='object').describe()
```

```python
# Checking for Duplicates
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
```

```python
# Checking for Outliers
numeric_cols = [
    'Open Price', 'Close Price', 'Daily High', 'Daily Low', 'Trading Volume',
    'GDP Growth (%)', 'Inflation Rate (%)', 'Unemployment Rate (%)',
    'Interest Rate (%)', 'Consumer Confidence Index', 'Government Debt (Billion USD)',
    'Corporate Profits (Billion USD)', 'Forex USD/EUR', 'Forex USD/JPY',
    'Crude Oil Price (USD per Barrel)', 'Gold Price (USD per Ounce)',
    'Real Estate Index', 'Retail Sales (Billion USD)',
    'Bankruptcy Rate (%)', 'Mergers & Acquisitions Deals',
    'Venture Capital Funding (Billion USD)', 'Consumer Spending (Billion USD)'
]

# Calculate IQR for numeric columns
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Flag outliers
outliers = ((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR)))
outlier_counts = outliers.sum()
outlier_counts = outlier_counts[outlier_counts > 0] 
outlier_counts
```

#### Full Summary Report

```python
### Data Cleaning Methods

'''Even though this dataset is largely clean, the following are common data cleaning steps that I could have applied to similar datasets:

# 1. Handling Missing Data
- **Fill missing values (`fillna`)**

# Fill numeric columns with median
df['Close Price'] = df['Close Price'].fillna(df['Close Price'].median())
df['Close Price'] = df['Close Price'].fillna(0)

# Remove rows with any missing value
df.dropna(inplace=True)

2. Removing Duplicate Rows

# Check for duplicates
df.duplicated().sum()

# Drop duplicate rows
df.drop_duplicates(inplace=True)
'''
```

## Feature Engineering and Metrics

```python
# Creating Stock Market Metrics

# Daily price change
df['Daily_Change'] = df['Close Price'] - df['Open Price']

# Daily return percentage
df['Daily_Return_%'] = (df['Close Price'] - df['Open Price']) / df['Open Price'] * 100

# Daily volatility (high - low)
df['Volatility'] = df['Daily High'] - df['Daily Low']

# Debt-to-GDP ratio (just an example; GDP column may need adjustment if it's growth %)
df['Debt_to_GDP'] = df['Government Debt (Billion USD)'] / (df['GDP Growth (%)'] + 1e-6)

# Corporate Profits / GDP
df['CorporateProfits_to_GDP'] = df['Corporate Profits (Billion USD)'] / (df['GDP Growth (%)'] + 1e-6)

# Retail Sales / Consumer Spending
df['RetailSales_to_ConsumerSpending'] = df['Retail Sales (Billion USD)'] / df['Consumer Spending (Billion USD)']

# 30-day rolling average of daily returns
df['DailyReturn_30d_MA'] = df['Daily_Return_%'].rolling(30).mean()

# 30-day rolling standard deviation (volatility)
df['Volatility_30d'] = df['Volatility'].rolling(30).std()

```

```python
df.head()
```

```python
# Creating Time based Features

# Extracting Year, Month, Day, Weekday
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Weekday'] = df['Date'].dt.day_name()
```

## Exploratory Data Analysis

```python
df.describe(include='all').T
```

```python
# Selecting numeric col 

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Distribution Visuals

plt.figure(figsize=(15, 8))
df[numeric_cols].hist(bins=40, figsize=(15, 12))
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmaps

plt.figure(figsize=(16,12))
sns.heatmap(df[numeric_cols].corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Heatmap of Financial & Economic Variables")
plt.show()
```

```python
# Time series visualizations

plt.figure(figsize=(14,6))
plt.plot(df["Date"], df["Close Price"])
plt.title("Stock Index Close Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()

plt.figure(figsize=(14,6))
plt.plot(df["Date"], df["Inflation Rate (%)"], label="Inflation")
plt.plot(df["Date"], df["Interest Rate (%)"], label="Interest Rates")
plt.legend()
plt.title("Inflation vs Interest Rates Over Time")
plt.show()
```

```python
# Rolling stats

df["Close_30d_MA"] = df["Close Price"].rolling(window=30).mean()
df["Close_30d_Volatility"] = df["Close Price"].rolling(window=30).std()

plt.figure(figsize=(14,6))
plt.plot(df["Date"], df["Close Price"], label="Close Price", alpha=0.5)
plt.plot(df["Date"], df["Close_30d_MA"], label="30-day MA", linewidth=2)
plt.legend()
plt.title("30-Day Moving Average (MA)")
plt.show()
```

```python
# Relationship between stocks and macro factors

# Inflation vs market
sns.scatterplot(data=df, x="Inflation Rate (%)", y="Close Price", alpha=0.4)
plt.title("Inflation vs Stock Market")
plt.show()
```

```python
# Interest Rates vs market

sns.scatterplot(data=df, x="Interest Rate (%)", y="Close Price", alpha=0.4)
plt.title("Interest Rates vs Stock Market")
plt.show()
```

```python
# Pairplot

sns.pairplot(df[["Close Price", "GDP Growth (%)", "Inflation Rate (%)", 
                 "Interest Rate (%)", "Corporate Profits (Billion USD)"]],
             diag_kind="kde")
plt.show()
```
